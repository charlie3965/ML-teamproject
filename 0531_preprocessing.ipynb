```python
import pandas as pd
```

```python
dfTemp1 = pd.read_csv('petfinder-adoption-prediction/train/train.csv')
dfTemp2 = pd.read_csv('petfinder-adoption-prediction/test/test.csv')
```

```python
df = pd.concat([dfTemp1,dfTemp2])
df
```

```python
df = df.loc[df['Quantity']==1]
df = df.drop('Quantity', axis=1)
df = df.reset_index(drop=True)
df
```

```python
df = df.drop('RescuerID', axis=1)
df = df.drop('PetID', axis=1)
df
```

```python
df['Name'].isnull().value_counts()
```

```python
df['Name'] = df['Name'].fillna("No Name No Name")
df['Name'].isnull().value_counts()
```

```python
df['Gender'].value_counts()
```

```python
breedMixed = []

for i in range(0, len(df)):
    if df['Breed1'][i]==307 and (df['Breed2'][i]==307 or df['Breed2'][i]==0):
        breedMixed.append(True)
    else:
        breedMixed.append(False)

df['BreedMixed'] = breedMixed
df = df.drop('Breed1', axis=1)
df = df.drop('Breed2', axis=1)
df
```

```python
colorMixed = []

for i in range(0, len(df)):
    if df['Color2'][i]!=0 or df['Color3'][i]!=0:
        colorMixed.append(False)
    else:
        colorMixed.append(True)

df['ColorMixed'] = colorMixed
df = df.drop('Color2', axis=1)
df = df.drop('Color3', axis=1)
df
```

```python
df['ColorMixed'].value_counts()
```
